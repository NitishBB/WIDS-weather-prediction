# Machine Learning 
Machine learning is a method of data analysis that automates analytical model building. 

Using algorithms that iteratively learn from data, machine learning allows computers to find hidden insights without being explicitly programmed where to look.

**Types of machine learning problems**:

On basis of the nature of the learning “signal” or “feedback” available to a learning system

* Supervised learning: The model or algorithm is presented with example inputs and their desired outputs and then finding patterns and connections between the input and the output.

 Some real-life examples are:
Image Classification,Market Prediction/Regression
* Unsupervised learning: No labels are given to the learning algorithm, leaving it on its own to find structure in its input. It is used for clustering population in different groups.
* Semi-supervised learning: Problems where you have a large amount of input data and only some of the data is labeled, are called semi-supervised learning problems. These problems sit in between both supervised and unsupervised learning. 
* Reinforcement learning: A computer program interacts with a dynamic environment in which it must perform a certain goal (such as driving a vehicle or playing a game against an opponent)

**Two most common use cases of Supervised learning are:**

Classification: Inputs are divided into two or more classes, and the learner must produce a model that assigns unseen inputs to one or more (multi-label classification) of these classes and predicting whether or not something belongs to a particular class. 

Regression: It is also a supervised learning problem, that predicts a numeric value and outputs are continuous rather than discrete. For example, predicting the stock prices using historical data

Density estimation:

# **Linear regression**

```python
df1 = df.copy()
```

**checking skewness of our feature variable**

```python
for i in ['Temperature (C)','target Humidity','Wind Speed (km/h)','Wind Bearing (degrees)','Visibility (km)','Pressure (millibars)']:
  sns.kdeplot( x= i,data= df1)
  plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(data):
  vif = pd.DataFrame()
  vif['variables'] = data.columns
  vif['vif_score']= [variance_inflation_factor(data.values,i ) for i in range(data.shape[1])]

  return (vif) 
```

```python
x = df1.iloc[:,:]
calc_vif(x)
```

# **Avoiding multicollinearity**
#using BoxCox transformation for humidity

```python
from scipy.special import boxcox1p
df1['target Humidity']= boxcox1p(df1['target Humidity'],4.5)
sns.kdeplot(x= 'target Humidity', data= df1)
plt.show()
```

```python
df1['Pressure (millibars)']= boxcox1p(df1['Pressure (millibars)'],-4)
sns.kdeplot(x= 'Pressure (millibars)', data= df1)
plt.show()
```

```python
x = df1.iloc[:,:]
calc_vif(x)
```

```python
from sklearn.model_selection import train_test_split
x = df1.drop('target Humidity',axis=1)
y = df1['target Humidity']
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size= 0.3,random_state = 101)
```

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(xtrain,ytrain)
```

```python
lr_pred = lr_model.predict(xtest)
```

**scaling our feature variable**

```python
from sklearn.metrics import r2_score,mean_squared_error
mse = mean_squared_error(ytest,lr_pred)
lr_r2 = r2_score(ytest,lr_pred)
print('mse: ', mse)
print('r2:', lr_r2)
```

# **Neural network**

```python
df4 = df.copy()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Activation
```

```python
x = df3.drop('target Humidity',axis=1).values
y = df3['target Humidity'].values
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size= 0.3,random_state = 101)
```

```python
model =Sequential()
model.add(Dense(30,activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(30,activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(30,activation = 'relu'))

model.add(Dense(15,activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(27,activation = 'relu'))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(xtrain,ytrain,epochs = 30,verbose = 1,validation_data =(xtest,ytest),batch_size = 64)
```

```python
lossdf= pd.DataFrame(model.history.history)
lossdf.plot()
```

```python
ypred = model.predict(xtest)
```

```python
ytest
```

```python
pred = list()
for i in range(len(ypred)):
  pred.append(ypred[i].max())
```

```python
pred = np.array(pred)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
mse = mean_squared_error(ytest,pred)
r2 = r2_score(ytest,pred)
print('mse: ', mse)
print('r2:', r2)
```
